### Simple Slicing

### Subsetting with loc and iloc

Used to slice rows and columns as per required **[row, column]**

- loc is used to index using labels or integers - searches by index
- iloc is used for index using integers - searches by position

```python
trips_data.head()
```

```python
trips_data.loc[5,'duration']
```

```python
trips_data.loc[5]
```

```python
trips_data.iloc[5]
```

```python
trips_data.sort_values(['duration'], inplace=True)
```

```python
trips_data.head()
```

```python
trips_data.iloc[2:5,:]
```

```python
trips_data.loc[2:5,:]
```

```python
trips_data.sort_index(inplace=True)
```

### Subsetting with condition

Often we need to select data which satisfies the specific condition. 

operators used for conditions
- **==** equal to
- **!=** not equal to
- **>=** greater than equal to 
- **<=** less than equal to
- **~** negation

Can pass multiple condition with & (and) , | (or)

```python
trips_data['subscription_type']=='Customer'
```

```python
trips_data.loc[trips_data['subscription_type']=='Customer']
```

```python
trips_data.loc[trips_data['duration'] >= 1000]
```

```python
trips_data.loc[(trips_data['duration'] >= 1000)&(trips_data['subscription_type']=='Customer')]
```

### Difference between loc and simple slicing

.loc searches are faster on scale as they are optimised for locating based on the indices. Moreover complex conditions cannot be passed in simple slicing

```python
%%timeit -n 50
trips_data.loc[trips_data['subscription_type']=='Customer']
```

```python
%%timeit -n 50
trips_data[trips_data['subscription_type']=='Customer']
```

### Subsetting with masks

> Subsetting using isnull( ), isin( ) and other similar functions which give boolean results

```python
trips_data.isnull()
```

```python
trips_data.isnull().any(axis=1)
```

```python
trips_data.loc[trips_data.isnull().any(axis=1)]
```

```python
trips_data.loc[trips_data['zip_code'].isnull()]
```

```python
trips_data.loc[~(trips_data.isnull().any(axis=1))]
```

    - iloc is used to search using integers : works on position 
- subsetting with condition works with all the operators of ==, <=, >=, !=
- subsetting can be done with masks which give boolean results
